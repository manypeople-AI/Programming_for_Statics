First, use linspace to set the distance on the x horizontal axis, and create y_sin and y_cos by adding x to sin and cos, respectively. At this time, to express the position on the axis, the position is put in the position and the pi is expressed in the label using the r'$\pi$' expression. Now, plot each graph with a different color.


#<font color='green'> Explanation ends here!!!!</font>

# <font color='red'><<<**Finally Answer**>>></font>

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline


x = np.linspace(-np.pi, np.pi)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, color = "r")
plt.grid(True)

positions = [-np.pi,-np.pi/2,0, np.pi/2, np.pi]
labels = [r'$-\pi$', r'$-\pi/2$', 0, r'$+\pi/2$' ,r'$+\pi$']
plt.xticks(positions, labels)

positions2 = [-1,0,1]
labels2 = [-1,0,1]
plt.yticks(positions2, labels2)

plt.plot(x, y_cos,color="b")

plt.show()
````

After creating a graph in the same way as above, create an arrow using plt.annotate. In this case, the name is written using the same notation as above, and the position of the text is specified using xytext. Use arrowprops to set arrow properties.


#<font color='green'> Explanation ends here!!!!</font>

# <font color='red'><<<**Finally Answer**>>></font>

````python
import math
%matplotlib inline

plt.figure(figsize=(16,8))
x = np.linspace(-np.pi, np.pi)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin, color = "r",label = 'sine')
plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$', xy=((2*np.pi)/3, math.sqrt(3)/2),fontsize=14,ha='left',
            xytext=(0.84, 0.95), textcoords='axes fraction', arrowprops=dict(facecolor='black', width=1, shrink=0.1, headwidth=10)
            )
plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$', xy=((2*np.pi)/3, -1/2),fontsize=14,ha='left',
            xytext=(0.7, 0.15), textcoords='axes fraction', arrowprops=dict(facecolor='black', width=1, shrink=0.1, headwidth=10)
            )
plt.grid(True)

positions = [-np.pi,-np.pi/2,0, np.pi/2, np.pi]
labels = [r'$-\pi$', r'$-\pi/2$', 0, r'$+\pi/2$' ,r'$+\pi$']
plt.xticks(positions, labels)

positions2 = [-1,0,1]
labels2 = [-1,0,1]
plt.yticks(positions2, labels2)

plt.plot(x, y_cos,color="b",label = 'cosine')
plt.legend(loc='upper left',ncol=1,fontsize=15)

plt.show()
````

```

```

````python
s = np.linspace(0, 2 * np.pi, 30)
t = np.linspace(-0.5, 0.5, 30)
s, t = np.meshgrid(s, t)
````

### b> You can make x, y, and z from the meshgrids of (s,t)

````python
x = (1+(t/2)*np.cos(s/2))*np.cos(s)
y = (1+(t/2)*np.cos(s/2))*np.sin(s)
z = (t/2)*np.sin(s/2)
````

### c> Use np.ravel to flatten x,y and z

````python
x = np.ravel(x)
y = np.ravel(y)
z = np.ravel(z)
````

````python
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z, cmap='viridis');
````
