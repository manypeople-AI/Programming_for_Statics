````python
def ins_sort(inp_list):
  n=len(inp_list)
  for i in range(n):
    key=inp_list[i]
    j=i-1
    while (j >=0) & (key < inp_list[j]):
      inp_list[j+1] = inp_list[j]
      j = j-1
      inp_list[j+1] = key
  
  return inp_list
````

## Problem 2> [Video Record] (Merge sort) Using the merge sort in the classroom,you want to sort the following list:a = [6, 3, 7, 1, 8, 2, 5, 4, 9] Provide the step-by-step explanation on how the merge sort algorithm is working using the example of a = [6, 3, 7, 1, 8, 2, 5, 4, 9].



#<font color='green'> Explanation starts here!!!!</font>

Merge sort is a sorting algorithm using divide and conquer (Divide and Conquer) and recursive algorithms. In other words, it continues by dividing the given array into two until there is only one element left, then rearranging it in the order of size and merging it into an array of the original size.

image.png


#<font color='green'> Explanation ends here!!!!</font>

# <font color='red'><<<**Finally Answer**>>></font>

````python
def merge_sort(a): 
  n = len(a) 
  
  if n <= 1: 
    return a 

  mid = n // 2 
  g1 = merge_sort(a[:mid])
  g2 = merge_sort(a[mid:])
  result = [] 
  while g1 and g2: 
    if g1[0] < g2[0]: 
      result.append(g1.pop(0)) 
    else: 
      result.append(g2.pop(0)) 
      
  while g1: 
    result.append(g1.pop(0)) 
  while g2: 
    result.append(g2.pop(0)) 
  return result
````

````python
a = [6, 3, 7, 1, 8, 2, 5, 4, 9]
merge_sort(a)
````

## Problem 3> [Video Record] (Variation of quick sort) Similar to the idea of Quick Sort, you want to develop your own version of Quick Sort having two pivots for each round rather than using one pivot like the ordinary Quick Sort





#<font color='green'> Explanation starts here!!!!</font>


```
Step 1: Choose the last two elements in the (sub)array. Call these elements as the

Step 2: Divide the elements in the list into three groups:
Group 1: all elements in this group are less than or equal to the smaller pivot.
Group 2: all elements in this group are greater than one pivot and smaller than

Group 3: all elements in this group are greater than or equal to the bigger pivot.
We call this procedure partitioning. At this point, it doesn't matter what order the

Step 3: Conquer by recursively sorting the element in each group. Combine by

```

Quick sort algorithm, like merge sort, is a sorting algorithm using divide and conquer (Devide and Conquer) technique and recursive algorithm.
Unlike merge sort, which always splits based on the center, quicksort uses a random reference value called a pivot. There are several ways to choose the pivot value, but here we will use two fixed pivots. Looking at the code, first, set the smallest of the two last values ​​in the array received as input as pivot 1 and the larger one as pivot 2.
And define three groups to be separated as pivots as an empty list. Then, the for loop is executed. In group g1, elements smaller than the smaller pivot are put, in g2, values ​​between pivot1 and pivot2, and in g3, values ​​larger than pivot2. After repeating the process just again as a recursive function for the divided g1, g2, g3, and returning the sum of all the values ​​in order of size, you can see that the results are sorted and output as follows.


#<font color='green'> Explanation ends here!!!!</font>

# <font color='red'><<<**Finally Answer**>>></font>

````python
def dual_q_s(a):
  n = len(a)
  if n <=1:
    return a

  if a[n-2]>a[n-1]:
    pivot1 = a[n-1]
    pivot2 = a[n-2]
  else:
    pivot1 = a[n-2]
    pivot2 = a[n-1]

  g1 = []
  g2 = []
  g3 = []
  for i in range(n):
    if a[i]<=pivot1:
      g1.append(a[i])
    elif pivot1<a[i]<pivot2:
      g2.append(a[i])
    elif a[i]>=pivot2:
      g3.append(a[i])

  return dual_q_s(g1)+dual_q_s(g2)+dual_q_s(g3)
````

Failure rate in each level is defined by Failure rate at level n = (Number of players who didn’t clear the level n yet) / (Number of players who play level n) As a developer, you want to know which level is difficult. Write the function whose input is “N” and “stages”, and return is a list of sorted levels in decreasing order by failure rate. 






#<font color='green'> Explanation starts here!!!!</font>


```

a. “N” is the number of stages.
b. “stages” is a list of positive integers in {1, ... , N+1}
 * each element in “stages” means the current level that each player is playing
 * N+1 means that the player has cleared all levels.
c. If there are the same failure rate, the smaller indexed level comes first.
d. If there are no player who play the level, the failure rate of the level is defined
as 0.
Example:
N=5
stages = [2,1,2,6,2,4,3,3]
Then,
Failure rate of stage 1 is 1/8.
Failure rate of stage 2 is 3/7.
Failure rate of stage 3 is 2/4.

Failure rate of stage 5 is 0.

```

First, for each stage, use a for statement to find the stage failure rate and put it in the list answer. If no users have reached the stage, count ==0 is given as a condition to fail. Otherwise, the failure rate is the number of people who have not reached the stage divided by the number of people who have reached the stage, so the count is divided by the length as follows. is subtracted at the end to continue executing the for statement. If all the failure rates are found in this answer, it is sorted in descending order based on the failure rate. At this time, I used the sorted function, and since it first sorts in descending order based on the failure rate, the key value is sorted based on the failure rate using lambda as follows, and at this time, set reverse to true and output in descending order. And since what we want is the stage number in descending order of failure rate, if we extract only the 0th value from each element of the answer, put it in the answer, and return it, you can see that the output is the same as in the example below.


#<font color='green'> Explanation ends here!!!!</font>

# <font color='red'><<<**Finally Answer**>>></font>

````python
def solution(N, stages):
    answer = []
    length = len(stages)
    for i in range(1, N+1):
        count = stages.count(i)
        if count == 0:
            fail = 0
        else:
            fail = count / length
        answer.append((i, fail))
        length -= count


    answer = sorted(answer, key=lambda t: t[1], reverse=True)
    answer = [i[0] for i in answer]
    return answer
````

Now, the count_x_value function is finally constructed using these two functions. At this time, the initial start value is set to 0 and the end value is set to N-1.
````python
# 가장 왼쪽에 있는 타겟값 인덱스 찾는 함수
def first(array, target, start, end):
    if start > end:
        return None
    mid = (start + end) // 2
    # 가장 왼쪽에 있는 값을 발견했을 경우
    if (mid == 0 or array[mid-1] < target) and array[mid] == target:
        return mid
    elif array[mid] >= target:
        return first(array, target, start, mid-1)
    else:
        return first(array, target, mid+1, end)


# 가장 오른쪽에 있는 타겟값 인덱스 찾는 함수
def last(array, target, start, end):
    if start > end:
        return None
    mid = (start + end) // 2
    if (mid == N-1 or array[mid+1] > target) and array[mid] == target:
        return mid
    elif array[mid] > target:
        return last(array, target, start, mid-1)
    else:
        return last(array, target, mid+1, end)


def count_x_value(target, array):
    N = len(array)
    start = 0
    end = N-1

    first_idx = first(array, target, start, end)
    if first_idx == None:
        return 0
    last_idx = last(array, target, start, end)

    return last_idx - first_idx + 1
````

````python
def solution(participant, completion):
    answer = '' 
    p_dict = {} 
    c_dict = {} 
    
    for i in participant: 
        if i not in p_dict:
            p_dict[i] = 0 
        p_dict[i] += 1
    
    for i in completion: 
        if i not in c_dict:
            c_dict[i] = 0
        c_dict[i] += 1
    
    p_items = p_dict.items() 
    c_items = c_dict.items()

    answer = list(dict(p_items - c_items).keys())[0]
    return answer
````

First, it takes graphs to use as input and a starting ending point. Since it returns the shortest distance among distances, qu is given a starting point and d to determine the distance. At this time, if x is the same as the ending point, d is returned. If they are not the same, run a for statement for elements with x as a key in the graphs, and at this time, set the if statement so that it does not overlap the value in the list called done so that the length does not overlap. Also, when appending to qu, it is moved by one, so d+1 is appended, and a new p is appended to done so that the paths do not overlap behind. If you repeat this process, you can output d where x becomes the same as the ending, and if it is not the same as the ending point even though all the values ​​in qu are turned, 'infinity' is returned. Therefore, if you run the city_distance function by substituting the example given in the problem, the result is well output as follows.


#<font color='green'> Explanation ends here!!!!</font>

# <font color='red'><<<**Finally Answer**>>></font>

````python
def city_distance(graphs, starting, ending):
  qu=[]
  done=[]
  qu.append((starting, 0))
  done.append(starting)
  
  while qu:
    x, d = qu.pop(0)
    if x == ending:
      return d
    for p in graphs[x]:
      if p not in done:
        qu.append((p,d+1))
        done.append(p)

  return 'infinity'
````

## Problem 8 > (Solve maze) In solving maze algorithm, you have found the shortest road to the goal. Revise the Python code that the program finds the shortest road and the second shortest road (if exists) to the goal. (Here, the road means that you do not pass through the same place two times.)


#<font color='green'> Explanation starts here!!!!</font>


In question 8, an empty list called visited is created in addition to qu and done. Looking at the code how to use this visited, first, all elements of qu are checked using while qu. At this time, the 0th element is taken out from qu, and the last alphabet of the element is put into v to compare v and end value. If they are not the same, we take the elements of v from g one by one, put those not in done into qu in the form of x+done, and add them to done. While repeating this, a point occurs when v and end become the same. However, to find the second shortest path at this time, since done must not contain an end point, v is removed from done. Add an if statement. If it is not included before, append x to visited and remove v from done as mentioned before. After repeating this and saving all the shortest possible paths in visited, the while statement is exited. At this time, the element with index 0 of visite is the shortest distance, and the element with index 1 is the second shortest distance, so you can get the desired value by printing as follows.


#<font color='green'> Explanation ends here!!!!</font>

# <font color='red'><<<**Finally Answer**>>></font>

````python
def solve_maze(g, start, end):
  qu=[]
  done = []
  visited=[]
  qu.append(start)
  done.append(start)

  while qu:
    x = qu.pop(0)
    v = x[-1]
    if v == end:
      if v not in x[:-1]:
        visited.append(x)
        done.remove(v)
    for road in g[v]:
      if road not in done:
        qu.append(x+road)
        done.append(road)
  
  if len(visited)==1:
    print("최단거리는 {} 이며 두번째 최단거리는 없습니다.".format(visited))
  elif len(visited)>1:
    print("최단거리는 {}이며 두번째 최단거리는 {} 입니다.".format(visited[0],visited[1]))
  elif len(visited)==0:
    print("갈수 없습니다! 경로를 다시 지정해주세요!")
````
