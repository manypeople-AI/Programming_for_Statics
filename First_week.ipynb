## Problem 2> Write a Python script to generate and print a dictionary that contains a number (between 1 and n) in the form (x, x*x)


# <font color='green'> Explanation starts here!!!!</font>

Our goal is to obtain

```
Sample Dictionary(n=5):
Expected Output : {1: 1,2: 4,3: 9,4: 16,5: 25}
```


When adding a value to a dictionary, add it in the form dic_name[key]=value.

->Therefore, I first created an empty dictionary and thought about adding values ​​to the dictionary by looping from 1 to the given value.

# <font color='green'> Explanation ends here!!!!</font>

# <font color='red'><<<**Finally Answer**>>></font>

````python
def dic(n):
  output_dic = dict()
  for i in range(1,n+1):
    output_dic[i] = i*i
  print(output_dic)
````

````python
dic(5)
````

## Problem 3> (List and Loop) Write a Python program to print the even numbers from a given list.


# <font color='green'> Explanation starts here!!!!</font>

Our goal is to obtain

```
Sample List : [1,2,3,4,5,6,7,8,9]
Expected Result : [2,4,6,8]
```

First, in order to determine whether each element of the list is an even number, we need to know how to select only each element of the list.

```
> sample_list = [1,2,3,4,5,6,7]
> sample_list[0]
1
```
Therefore, the order of the list starts from 0, and in order to select the first element from the list as above, you need to write code like sample_list[0].


Therefore, it creates a new empty list called new_list, selects only even numbers from input_list, and uses append to new_list.

# <font color='green'> Explanation ends here!!!!</font>

# <font color='red'><<<**Finally Answer**>>></font>

````python
def even_list(input_list):
  new_list = []
  for i in range(len(input_list)):
    if input_list[i]%2==0:
      new_list.append(input_list[i])
  
  print(new_list)
````

````python
even_list([1,2,3,4,5,6,7,8,9])
````

## Problem 4> (Loop) Write a Python code that prints out the first n rows of Pascal's triangle.


# <font color='green'> Explanation starts here!!!!</font>

Our goal is to obtain

```
For example, for n=3, your output should be
[1]
[1,1]
[1,2,1]
```

Let's analyze the features of Pascal's triangle first.



image.png

What you see here is a Pascal's triangle.
1st row 1, 2nd row 11, 3rd row 1 2 1 4th row 1331
is composed. At this time, if you look at the rules, first, the second element of the third line, 2, is
It is the sum of 1 1 in the second row, and 3, the second element in the fourth row, is the first in the third row.
It is the sum of the element 1 and the second element 2. That is, the value of the i-th element of the n-th line is
It is the sum of the values ​​of the i-th element and the i-1th element of the n-1th row.

I tried to compose a function based on these characteristics.
The name of the function is pascal and it takes a number n as an argument.
The function is mainly composed of two for loops.
The first for loop is now responsible for the lines of the Pascal triangle.
The second for loop takes care of the elements in that line.

# <font color='green'> Explanation ends here!!!!</font>

# <font color='red'><<<**Finally Answer**>>></font>

````python
def pascal(n):
  list_pas = [1]
  print(list_pas)
  for i in range(1,n):
    list_new = [0 for m in range(i+1)]
    list_pas.append(0)
    for j in range(1,len(list_new)+1):
      if j>=2:
        list_new[j-1] = list_pas[j-2] + list_pas[j-1]
      else:
        list_new[j-1] = list_pas[j-1]
    list_pas = list_new
    print(list_new)
````

````python
pascal(5)
````

## Problem 5> (Loop) Write a Python program to find a pair with highest product from a given array of integers


# <font color='green'> Explanation starts here!!!!</font>

Our goal is to obtain

```
Original array : [1,2,3,4,7,0,8,4]
Maximum product pair is : (7,8)
Original array : [0,-1,-2,-4,5,0,-6]
Maximum product pair is : (-4,-6)
``` 

The product of each element is obtained one by one using the for statement, and the values ​​of the product are stored each time, and then the product elements of the one finally selected as the max value are obtained.


First, the max value is initialized to 0, compared with the product of the elements in the list, and finally the largest value is put into max, and the two values ​​used to create the max are put into max_pair.

# <font color='green'> Explanation ends here!!!!</font>

# <font color='red'><<<**Finally Answer**>>></font>

````python
def max_pair(array):
  max = 0
  for i in range(len(array)):
    for j in range(i+1,len(array)):
      a = array[i]*array[j]
      if a>max:
        max = a 
        max_pair = (array[i],array[j])
  print(max_pair)
````

````python
max_pair([0,-1,-2,-4,5,0,-6])
````

## Problem 5-2> (Algorithm) Write a python code to find the maximum value and the index of the maximum value in "my_num". 


# <font color='green'> Explanation starts here!!!!</font>

Our goal is to obtain

```
my_num = [3,5,10,7,2,4]
maximum value = 10
Index of the maximum value = 2

``` 

First, you can use the max function to find the maximum value in a list.

Second, you can use .index to find the position of a specific element in a list.



# <font color='green'> Explanation ends here!!!!</font>

# <font color='red'><<<**Finally Answer**>>></font>

````python
def list_max(my_num):
  a = max(my_num)
  print('maximum values : ', a)
  print('Index of the maximum values : ',my_num.index(a))
````

````python
my_num = [3,5,10,7,2,4]
list_max(my_num)
````


## Problem 6> (Algorithm) Given a list of integers with duplicate elements in it, write a Python code to generate another list, which contains only the duplicate elements. In simple words, the new list should contain the elements which appear more than one


# <font color='green'> Explanation starts here!!!!</font>

Our goal is to obtain

```
1. Consider that you are given a list of A = [Tom, Jerry, Mike, Tom, Mike]
2. Compare Tom with all other elements in A[1:]. --> add Tom to the list
3. Compare Mike with all other elements in A[3:]. --> add Mike to the list.
4. Do the same thing for Tom.
``` 
Hint : you can use 'set' to solve this problem.

First of all, two different elements can be compared by making FOR LOOP into two main parts. Therefore, if two elements in different order are the same, it is a duplicate element, so a duplicate_list containing duplicate elements was created and added using append. Or using the set that the professor gave
Duplicate values ​​can be added using s.add.

# <font color='green'> Explanation ends here!!!!</font>

# <font color='red'><<<**Finally Answer 1**>>></font>

````python
def duplicate(my_list):
  duplicate_list = []
  for i in range(len(my_list)):
    for j in range(i+1,len(my_list)):
      if my_list[i] ==my_list[j]:
        duplicate_list.append(my_list[i])
  print(duplicate_list)
````

````python
my_list = ['Tom', 'Jerry', 'Mike', 'Tom', 'Mike']
duplicate(my_list)
````

````python
def duplicate_2(my_list):
  S=set()
  for i in range(len(my_list)):
    for j in range(i+1, len(my_list)):
      if my_list[i] == my_list[j]:
        S.add(my_list[i])
  print(list(S))
````
