## Problem 1>  Write a list comprehension statement to generate a list of all pairs of odd positive integer values less than 10 where the first value is less than the second value.


#<font color='green'> Explanation starts here!!!!</font>

Our goal is to obtain

```
[(1,3),(1,5),(1,7),(1,9),(3,5),(3,7),(3,9),(5,7),(5,9),(7,9)]
```


Problem 1 is a problem of writing a list comprehension statement that creates a list containing two odd integers less than 10 as follows, paired in order so that the smaller number comes first. Therefore, the for loop is returned for each i located in the first and j located in the second using the nested for loop as follows. At this time, since i must be an odd number, put i%2!=0 as a condition in the if statement, and for j, i+2*j must be less than 10, so the condition is set as follows. Also set the range to (1,5) .  Therefore, if you write a list comprehension statement as follows, you can check that the result is good.


#<font color='green'> Explanation ends here!!!!</font>

# <font color='red'><<<**Finally Answer**>>></font>

````python
odd_pair = [(i,i+2*j) for i in range(1,10) if i%2!=0 for j in range(1,5) if i+2*j<10]
odd_pair
````

## Problem 3> Using filter function in Python, write the code to extract the outliers in the sample.


#<font color='green'> Explanation starts here!!!!</font>

In this problem, we define outliers as data points that lie more than two standard deviations from the sample mean
```
sample = [10, 8, 10, 8, 2, 7, 9, 3, 34, 9, 5, 9, 25]
```

Problem #3 is to remove outliers using the filter function. As shown in the problem, there is no set standard for determining outliers. Therefore, it is necessary to extract outliers through the criteria of the given outliers in the problem. Since the standard of outlier is specified as “we define outliers as data points that lie more than two standard deviations from the sample mean” in the problem, the mean and standard deviation of the list given as a sample are first calculated as the mean and std of numpy as follows. Calculate using it. Then, write the outlier function to return only the value greater than the mean minus the double std and smaller than the addition as written below. Then, if you use the filter to output the list as a list, you can see that the result is printed well.

#<font color='green'> Explanation ends here!!!!</font>

# <font color='red'><<<**Finally Answer**>>></font>

````python
import numpy as np

sample = [10, 8, 10, 8, 2, 7, 9, 3, 34, 9, 5, 9, 25]
mean = np.mean(sample)
std = np.std(sample)

def outlier_(x):
  return mean-2*std<x<mean+2*std

list(filter(outlier_,sample))
````

## Problem 4> (filter) Using filter function in Python, write the code to extract the palindrome in “words”as list.

#<font color='green'> Explanation starts here!!!!</font>

You are given the following words.

```
words = ("filter", "Ana", "hello", "world", "madam", "racecar")
```

Problem #4 is a problem of extracting only the strings that are palindrome among the strings in words using the filter function. To solve this problem, first write a function that can pick out palindrome. The Palindrome function was written with reference to the function previously written in HW2. Therefore, if you write a function to return only the palindromes as shown below and then use the filter to extract the palindromes into a list, you can see that it is well extracted as follows.

#<font color='green'> Explanation ends here!!!!</font>

# <font color='red'><<<**Finally Answer**>>></font>

````python
def palindrome(string):
  import math
  string = string.lower()
  characters = "'!?,.~@ "
  for x in range(len(characters)):
    string = string.replace(characters[x],"")
  for i in range(math.floor(len(string)/2)):
    if string[i]!=string[len(string)-(i+1)]:
      return 0
  return string


words = ("filter", "Ana", "hello", "world", "madam", "racecar")
list(filter(palindrome,words))
````

```
matrix_ = [[3,2,5],
 [4,1,7],
 [10,11,12]]
```
You want to flatten the matrix into 1-dim list so that it becomes
flat = [3,2,5,4,1,7,10,11,12]

Problem #5 is a problem of flattening a two-dimensional matrix using list comprehensions. A matrix is given as a list of 2d as follows. To output this as a one-dimensional list, I wrote a list comprehension as follows. First, I wrote matrix[i][j] as follows to get each element in the matrix in turn. Then, because i corresponds to the row, the matrix outputs the shape using np.array, and among them, the 0th corresponding to the row is put into the range, and the for statement is returned in the row part. Similarly, j corresponds to the column, so in the shape of the array Put the first column corresponding to the range and run the for statement in the column part, and the elements are put into the list one by one through i and j. Therefore, if you finally print the output, you can check that the matrix is printed well as in the sample above.

#<font color='green'> Explanation ends here!!!!</font>

# <font color='red'><<<**Finally Answer**>>></font>

````python
matrix_ = [[3,2,5],
 [4,1,7],
 [10,11,12]]

flat = [matrix_[i][j] for i in range(np.array(matrix_).shape[0]) for j in range(np.array(matrix_).shape[1])]
flat
````
